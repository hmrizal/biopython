```python
# import library
import numpy as np
import pandas as pd
from Bio import Entrez
```

```python
# enter e-mail to be recognized by Entrez
Entrez.email = 'learnbiopython@gmail.com'
```

```python
# list of the scientific names
terms = ['Bos gaurus', 'Antelope cervicapra', 'Gazella bennettii', 'Boselaphus tragocamelus', 'Canis lupus', 'Panthera leo', 'Elephas maximus', 'Equus africanus', 'Panthera pardus', 'Cervus canadensis', 'Pavo cristatus', 'Grus leucogeranus', 'Vulpes vulpes', 'Rhinoceros unicornis', 'Panthera Tigris', 'Crocodylus palustris', 'Gavialis gangeticus', 'Equus caballus', 'Equus quagga', 'Babalus bubalis', 'Sus scrofa', 'Camelus dromedaries', 'Giraffa camelopardalis ', 'Hemidactylus flaviviridis', 'Hippopotamus amphibius', 'Macaca mulatta', 'Canis lupus', 'Felis domesticus', 'Acinonyx jubatus', 'Rattus rattus', 'Mus musculus', 'Oryctolagus cuniculus', 'Bubo virginianus', 'Passer domesticus', 'Corvus splendens', 'Acridotheres tristis', 'Psittacula eupatria', 'Molpastes cafer', 'Eudynamis scolopaccus', 'Columba livia', 'Naja naja', 'Ophiophagus hannah', 'Hydrophiinae ', 'Python molurus', 'Ptyas mucosa']
```

```python
# 1. Correct the wrong scientific names using Entrez

name = []
animalName = []
correction = []

for term in terms:
  record = Entrez.read(Entrez.espell(db='taxonomy', term=term))
  name.append(record['Query'])
  animalName.append(record['CorrectedQuery'].lower())

  if (record['Query'].lower() != record['CorrectedQuery'].lower()):
    correction.append(record['CorrectedQuery'].lower())
  else:
    correction.append('-')
  
# Create dataframe
df = pd.DataFrame(list(zip(name, correction)), columns=["question","correction"])
df.index = np.arange(1, len(df) + 1)
df['status'] = df['correction'] == '-'

df
```

```python
# determining the name of the file
file_name = 'correction.xlsx'
  
# saving the excel
df.to_excel(file_name)
print('DataFrame is written to Excel File successfully.')
```

```python
# 2. Get common name from scientific name using Entrez

scientificName = []
commonName = []

for animal in animalName:
  search = Entrez.read(Entrez.esearch(db='taxonomy', term=animal))
  if (search['IdList']):
    fetch = Entrez.read(Entrez.efetch(db='taxonomy', id=search['IdList']))
    
    if 'GenbankCommonName' in fetch[0]['OtherNames']:
      if (fetch[0]['OtherNames']['GenbankCommonName'] != None):
        scientificName.append(fetch[0]['ScientificName'])
        commonName.append(fetch[0]['OtherNames']['GenbankCommonName'])


# Create dataframe
df2 = pd.DataFrame(list(zip(scientificName, commonName)), columns=["ScientificName","CommonName"])
df2.index = np.arange(1, len(df2) + 1)
df2
```

```python
# determining the name of the file
file2_name = 'name.xlsx'
  
# saving the excel
df2.to_excel(file2_name)
print('DataFrame is written to Excel File successfully.')
```

```python
# 3.1. Get taxonomy rank of Panthera tigris

searchTiger = Entrez.read(Entrez.esearch(db='taxonomy', term='panthera tigris'))
fetchTiger = Entrez.read(Entrez.efetch(db='taxonomy', id=searchTiger['IdList']))

print("Rank: ", fetchTiger[0]['Rank'])
```

```python
# 3.2. get lineage of Panthera tigris

print(fetchTiger[0]['Lineage'])

lineageTiger = fetchTiger[0]['LineageEx']

for item in lineageTiger:
  print(item['Rank'], ": ", item['ScientificName'])
```
